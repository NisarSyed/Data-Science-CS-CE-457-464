### Part 2


```python
df = pd.read_csv("formatted_580SurveyCleanup2.csv", )
df.head()
```

Since there are no values of `Access Code` and `Email Address`, these columns will not contribute to our analysis and therefore, they can be dropped from the dataframe.


```python
df_part2 = df.drop(['Access Code', 'Email Address'], axis=1)
df_part2.head()
```

The `Region` and `City / Town / District` columns have a few missing values and they can be filled in with the a default value of **Unknown**


```python
df_part2.fillna({'Region':'Unknown'}, inplace=True)
df_part2.fillna({'City / Town / District': 'Unknown'}, inplace=True)

df_part2.isnull().sum()
```

```python
df_part2.Q3.unique()
```

`Q3` has a few values that don't follow the format in the column. These values are handled individually.


```python
from datetime import datetime

df_part2.replace({"Q3":"2 3"}, "23", inplace=True)
df_part2.replace({"Q3":"1993"}, str(datetime.now().year - 1993), inplace=True)

df_part2.Q3.unique()
```

```python
df_part2.Q4.unique()
```

`Q4` has a missing value and some out of format entries. This column has only numerical entries therefore, the missing value is filled in by a default **0**. Later on, it will be updated with the mean of the column once datatype is also handled. Out of format values are handled individually.


```python
df_part2.fillna({'Q4':'0'}, inplace=True)
df_part2.replace({'Q4':'5.2'}, "52", inplace=True)
df_part2.replace({'Q4':'6l'}, "61", inplace=True)
df_part2.replace({'Q4':'7O'}, "70", inplace=True)

df_part2.Q4.unique()
```

```python
df_part2.Q5.unique()
```

`Q5` has missing values, values with extra whitespace, and repeated values. Missing value is filled with a default **Unknown**. Other values are handled individually.


```python
df_part2.fillna({'Q5':'Unknown'}, inplace=True)
df_part2.Q5 = df_part2.Q5.str.strip()
df_part2.Q5 = df_part2.Q5.replace({
    'US': 'United States',
    'USA': 'United States',
    'United States ': 'United States',
    'United States ; Pakistan': 'United States, Pakistan',
    'china': 'China',
    'Uzbekistan ': 'Uzbekistan',
    'India ': 'India',
    'Nigeria ': 'Nigeria',
    'pakistan': 'Pakistan',
})

df_part2.Q5.unique() 
```

```python
df_part2.Q6.unique()
```

`Q6` has missing values, values with extra whitespace, and repeated values. Missing value is filled with a default **Unknown**. Other values are handled individually.


```python
df_part2.fillna({'Q6':'Unknown'}, inplace=True)
df_part2.Q6 = df_part2.Q6.str.strip()
df_part2.Q6 = df_part2.Q6.replace({
    'economics': 'Economics',
    'Economic': 'Economics',
    'Economics with Finance': 'Economics & Finance',
    'Yes': 'Unknown',
    'BIT': 'Business Information Technology',
    'Cse': 'Computer Science & Engineering',
    'Accounting and ISOM': 'Accounting & ISOM',
    'Computer science': 'Computer Science',
    'Civil engineering': 'Civil Engineering',
    'ECE': 'Electrical & Computer Engineering',
    'Bachelor of Engineering': 'Engineering',
    'Bachelor of engineering': 'Engineering',
    'B.Tech in CS': 'Computer Science',
    'BS accounting': 'Accounting',
    'Criminology Law and Soci': 'Criminology, Law & Society',
})

df_part2.Q6.unique() 
```

```python
df_part2.Q7.unique()
```

`Q6` has only one missing value, which is filled with a default **Unknown**.


```python
df_part2.fillna({'Q7':'Unknown'}, inplace=True)
df_part2.Q7.unique()
```

```python
df_part2.Q8.unique()
```

`Q8` has missing values which are filled with a default **Unknown**. A new category **Both** is created for users that use both type of laptops.


```python
df_part2.fillna({'Q8':'Unknown'}, inplace=True)
df_part2.Q8 = df_part2.Q8.replace({
    'Microsoft/WindowsApple/MacBook': 'Both',
})

df_part2.Q8.unique()
```

`Q9`, `Q10`, `Q11`, `Q12` have missing values which are filled in with a default **Unknown**.


```python
df_part2.fillna({'Q9':'Unknown', 'Q10':'Unknown', 'Q11':'Unknown', 'Q12':'Unknown'}, inplace=True)
df_part2.isnull().sum()
```

Datatypes are changed for the relevant columns and the **0** value from `Q4` has been updated with the mean value of female height.


```python
df_part2.Q1 = df_part2.Q1.astype(object)
df_part2.Q3 = df_part2.Q3.astype('int64')
df_part2.Q4 = df_part2.Q4.astype('int64')

fem_height_mean = df_part2[df_part2.Q2 == "Female"].Q4.mean()
df_part2.replace({'Q4':0}, int(fem_height_mean), inplace=True)

df_part2.info()
```

```python
df_part2.to_csv("cleaned_580SurveyCleanup2.csv", index=False)
```
